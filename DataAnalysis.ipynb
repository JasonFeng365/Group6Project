# 1. MNIST

Make sure to install cv2 and tensorflow libraries before importing!!!

```python
#Import all the libraries
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import sklearn
```

```python
# import our dataset using tensorflow
mnist = tf.keras.datasets.mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
X = np.append(x_train, x_test, axis=0)
Y = np.append(y_train, y_test, axis=0)
print(X.shape)
print(Y.shape)
```

```python
# Define constants
labels="0123456789+-x/()"
```

### Mean values per label

```python
mnistTotalSums = [[[0 for col in range(28)] for row in range(28)] for label in range(10)]
occurrences = [0]*10
for image, label in zip(X.tolist(), Y.tolist()):
    occurrences[label]+=1
    for r in range(28):
        for c in range(28):
            mnistTotalSums[label][r][c] += image[r][c]
```

```python
means = [[[0 for col in range(28)] for row in range(28)] for label in range(10)]
for label in range(10):
    for r in range(28):
        for c in range(28):
            means[label][r][c] = mnistTotalSums[label][r][c]//occurrences[label]
```

```python
for label in range(10):
    print(labels[label])
    plt.imshow(means[label], cmap=plt.cm.binary)
    plt.show()
```

# 2. Handwritten dataset

```python
# Jason's 20 x 16 dataset

dim = 28
    
def getHandwrittenData(path="handwritten_training_data.png", verbose=False, displayOrig=True, factor=1):
    img = cv2.imread(path)[:,:,0]

    if displayOrig:
        temp = np.invert(np.array([img]))
        plt.imshow(temp[0], cmap=plt.cm.binary)
        plt.show()
    
    resInput = [None]*16
    for row in range(16):
        cur = [None]*20
        for col in range(20):
            sub = img[row*29+1:row*29+29, col*29+1:col*29+29]
            sub = np.invert(np.array([sub]))
    
            cur[col]=sub[0]
            if verbose:
                plt.imshow(sub[0], cmap=plt.cm.binary)
                plt.show()
        resInput[row] = cur

    resOutput = [None]*(20*16)
    for i in range(16):
        for idx in range(i*20, i*20+20):
            resOutput[idx]=i
    resInput = resInput * factor
    resOutput = resOutput * factor
    return resInput, resOutput
```

```python
X, Y = getHandwrittenData()
temp = []
for arr in X: temp+=arr
X = temp
# Y = [arr.index(1) for arr in Y]
# for arr in Y: temp+=arr
# Y = temp
```

### Mean values per label

```python
handwrittenTotalSums = [[[0 for col in range(28)] for row in range(28)] for label in range(16)]
occurrences = [0]*16
for image, label in zip(X, Y):
    occurrences[label]+=1
    for r in range(28):
        for c in range(28):
            handwrittenTotalSums[label][r][c] += image[r][c]
```

```python
means = [[[0 for col in range(28)] for row in range(28)] for label in range(16)]
for label in range(16):
    for r in range(28):
        for c in range(28):
            means[label][r][c] = handwrittenTotalSums[label][r][c]//occurrences[label]
```

```python
for label in range(16):
    print(labels[label])
    plt.imshow(means[label], cmap=plt.cm.binary)
    plt.show()
```
